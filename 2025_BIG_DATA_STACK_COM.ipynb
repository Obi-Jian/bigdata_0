```python
#3A riferimento foglio N1 su excel
#rahul

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# Dati
x = np.array([-4, -3, -2, -1, 0, 1, 2, 3, 4])
y = np.abs(x) #valore assoluto absolute abs

# Creazione del grafico
plt.figure(figsize=(8, 6))  # Imposta le dimensioni del grafico
plt.plot(x, y, marker='o', linestyle='-', color='b', label='y = |x|')  # Disegna la unzione
plt.title("Grafico della funzione y = |x|")  # Titolo del grafico
plt.xlabel("x")  # Etichetta dell'asse x
plt.ylabel("y")  # Etichetta dell'asse y
plt.grid(True)  # Aggiungi una griglia
plt.axhline(0, color='black', linewidth=0.5)  #Linea orizzontale per y =0
plt.axvline(0, color='black', linewidth=0.5)  #Linea verticale per x = 0
plt.legend()
plt.show()  #   grafico
coefficiente_pearson, _ = pearsonr(x, y)
print()#spazio
print(f"Il coefficiente di Pearson è: {coefficiente_pearson}")
print("Questo indica che non c'è alcuna correlazione lineare tra le due variabili.")
```

## TASK 3 B

A partire dal dataset della Tabella di Mollier, implementare
es. su Notebook, usando sia Excel che Python, graficando i
dati (T vs. Vl
) e calcolando i coeff. di Pearson e Spearman

```python
#3B riferimento foglio N2 su excel
#grampa
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr, spearmanr, linregress

# Dati della Tabella di Mollier
T = np.array([0, 5, 10, 15, 20, 30, 40, 60, 80, 100, 150, 200, 250, 300, 374])  # Temperatura in °C
v1 = np.array([0.0179, 0.0179, 0.0179, 0.0179, 0.0179, 0.0180, 0.0180, 0.0183, 0.0185, 0.0187, 0.0196, 0.2070, 0.0224, 0.0252, 0.0570])  # Volume specifico v1

# Calcolo dei coefficienti di correlazione
pearson_corr, _ = pearsonr(T, v1)
spearman_corr, _ = spearmanr(T, v1)

# Calcolo della retta di regressione lineare
slope, intercept, r_value, p_value, std_err = linregress(T, v1)
line = slope * T + intercept  # Retta di regressione

# Creazione del grafico
plt.figure(figsize=(10, 6))
plt.scatter(T, v1, color='blue', label="Dati")
plt.plot(T, line, color='red', label="Retta di regressione")  # Aggiungi la linea di regressione
plt.title("Grafico a dispersione: T vs. v₁")
plt.xlabel("Temperatura (°C)")
plt.ylabel("Volume specifico v₁ (m³/kg)")
plt.grid(True)
plt.legend()

# Aggiungi annotazioni per i coefficienti di correlazione
plt.text(50, 0.15, f"Pearson: {pearson_corr:.3f}", fontsize=12, color='red')
plt.text(50, 0.13, f"Spearman: {spearman_corr:.3f}", fontsize=12, color='green')

# Imposta i limiti degli assi per corrispondere al grafico fornito
plt.xlim(0, 400)  # Asse x da 0 a 400 °C
plt.ylim(0, 0.25)  # Asse y da 0 a 0.25 m³/kg

# Mostra il grafico
plt.show()

# Stampa dei coefficienti di correlazione
print()#spazio
print(f"Coefficiente di Pearson: {pearson_corr:.3f}")
print(f"Coefficiente di Spearman: {spearman_corr:.3f}")
```

Implementare l'es. del traffico milanese su

```python
#3C Riferimento foglio N3 su excel
#grampa
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Dati presi dalla tabella
traffic_data = {
    'Day': ['lun', 'mar', 'mer', 'gio', 'ven', 'sab', 'dom'] * 4,
    'Traffico': [100, 90, 80, 80, 90, 60, 40, 105, 95, 80, 85, 95, 60, 35,
                 100, 90, 75, 80, 90, 65, 40, 105, 90, 80, 85, 90, 55, 35]
}
df = pd.DataFrame(traffic_data)

df['Index'] = range(1, len(df) + 1)

# Grafico traffico a Milano
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(df['Index'], df['Traffico'], 'r-', label='Traffico')
plt.xlabel('Giorni')
plt.ylabel('Indice Traffico')
plt.title('Traffico a Milano')
plt.legend()
plt.grid(True)

# Calcolo dell'autocorrelazione
lags= range(1, 27)
#lags = range(1, len(df)//2)
autocorr = [df['Traffico'].autocorr(lag=lag) for lag in lags]

# Grafico autocorrelazione
plt.subplot(1, 2, 2)
plt.plot(lags, autocorr, 'bo-', label='Autocorrelazione')
plt.xlabel('Shift in Mesi')
plt.ylabel('Coeff. Autocorrelazione')
plt.title('Autocorrelazione')
plt.axhline(0, color='gray', linestyle='--')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
#Grafico 1 popolazione italiana random
#rahul
import numpy as np
import matplotlib.pyplot as plt

# Generazione dati casuali
np.random.seed(42) #se metti 42 nella parentesi ogni volta che esegui il codice otterrai gli stessi valori casuali.
n = 500  # Numero di individui

# Altezza in metri (tra 0.5m per un bambino e 2.2m per un adulto molto alto)
heights = np.random.uniform(0.5, 2.2, n)

# Peso in kg basato su un rapporto realistico con l'altezza
weights = heights ** 2 * np.random.uniform(18, 40, n)

# Calcolo del BMI
bmi = weights / (heights ** 2)

# Classificazione in categorie BMI
categories = np.where(bmi < 25, "normale", np.where(bmi < 30, "sovrappeso", "obeso"))
colors = {"normale": "red", "sovrappeso": "green", "obeso": "blue"}

# Creazione del grafico
plt.figure(figsize=(8,5))

for category, color in colors.items():
    mask = categories == category
    plt.scatter(weights[mask], heights[mask] * 100, c=color, label=category, alpha=0.6, s=10, marker='.')

# Impostazioni del grafico
plt.xlabel("Peso (kg)")
plt.ylabel("Altezza (cm)")
plt.title("Distribuzione della popolazione italiana tramite BMI")
plt.legend(markerscale=2, loc="upper right")
plt.grid(False)
plt.show()
```

Nel codice abbiamo generato casualmente i dati, comunque utilizzando il comando "np.random.seed(42)" ogni volta che il codice viene eseguito i dati non variano.

```python
#5B
# se non runnate il package installato appositamente per tale risorsa(pip install...) darà errore l'esecuzione di questo codice.
import trueskill

# Impostiamo l'ambiente TrueSkill
env = trueskill.TrueSkill(draw_probability=0)  # Nessun pareggio per semplicità

# Creiamo alcuni giocatori con punteggi iniziali
players = {
    "Alice": env.create_rating(),
    "Bob": env.create_rating(),
    "Charlie": env.create_rating(),
    "David": env.create_rating()
}

def simulate_match(player1, player2, winner):
    """Simula una partita tra due giocatori e aggiorna i punteggi."""
    ratings = [players[player1], players[player2]]

    if winner == player1:
        ranks = [0, 1]  # Il primo giocatore vince
    else:
        ranks = [1, 0]  # Il secondo giocatore vince

    new_ratings = env.rate([(ratings[0],), (ratings[1],)], ranks)

    players[player1], players[player2] = new_ratings[0][0], new_ratings[1][0]

def simulate_team_match(team1, team2, winner):
    """Simula una partita tra squadre e aggiorna i punteggi."""
    team1_ratings = [players[p] for p in team1]
    team2_ratings = [players[p] for p in team2]

    if winner == "team1":
        ranks = [0, 1]  # La prima squadra vince
    else:
        ranks = [1, 0]  # La seconda squadra vince

    new_ratings = env.rate([team1_ratings, team2_ratings], ranks)

    for i, p in enumerate(team1):
        players[p] = new_ratings[0][i]
    for i, p in enumerate(team2):
        players[p] = new_ratings[1][i]

# Simuliamo alcune partite
simulate_match("Alice", "Bob", "Alice")
simulate_match("Charlie", "David", "Charlie")
simulate_team_match(["Alice", "Charlie"], ["Bob", "David"], "team1")

# Stampiamo i punteggi aggiornati
print("Punteggi aggiornati:")
for player, rating in players.items():
    print(f"{player}: mu={rating.mu:.2f}, sigma={rating.sigma:.2f}")
```

```python
#5C
#se non runnate il package installato appositamente per tale risorsa(pip install...) darà errore l'esecuzione di questo codice.
import wikipediaapi #libreria Python che consente di interagire facilmente con i contenuti di Wikipedia
import numpy as np

def get_wikipedia_stats(name):
    """Recupera statistiche dalla pagina Wikipedia di un personaggio."""
    wiki = wikipediaapi.Wikipedia(user_agent='WikipediaRankingBot', language='en')
    page = wiki.page(name)

    if not page.exists():
        return None

    return {
        'name': name,
        'length': len(page.text),  # Numero di caratteri nella pagina
        'num_langs': len(page.langlinks),  # Numero di lingue disponibili
        'num_links': len(page.links)  # Numero di link interni
    }

def normalize_scores(scores):
    """Normalizza i punteggi tra 0 e 1 per evitare errori di divisione."""
    min_val, max_val = min(scores), max(scores)
    if max_val == min_val:
        return [0.5] * len(scores)  # Se tutti uguali, assegna 0.5 a tutti
    return [(s - min_val) / (max_val - min_val) for s in scores]

def compute_ranking(names):
    """Calcola il ranking basato sui dati Wikipedia."""
    data = [get_wikipedia_stats(name) for name in names]
    data = [d for d in data if d is not None]

    if not data:
        print("Nessun dato trovato.")
        return

    # Estrai i dati normalizzati
    lengths = normalize_scores([d['length'] for d in data])
    num_langs = normalize_scores([d['num_langs'] for d in data])
    num_links = normalize_scores([d['num_links'] for d in data])

    # Pesi assegnati ai fattori
    scores = [(0.4 * l + 0.3 * n + 0.3 * m) for l, n, m in zip(lengths, num_langs, num_links)]
    ranking = sorted(zip([d['name'] for d in data], scores), key=lambda x: x[1], reverse=True)

    print("Classifica Wikipedia:")
    for rank, (name, score) in enumerate(ranking, 1):
        print(f"{rank}. {name} - Punteggio: {score:.2f}")

names = ["Romulo", "Giulio Cesare", "Dante Alighieri", "Leonardo Da Vinci", "Benito Mussolini",
         "Umberto II Re d'Italia", "Rita Levi Montalcini", "Alessandro Manzoni", "Mario Draghi"]
compute_ranking(names)
```

```python
import itertools
import random
import matplotlib.pyplot as plt
from collections import Counter

# Definiamo gli elementi da permutare
elements = [1, 2, 3, 4]

# Generiamo tutte le possibili permutazioni
all_permutations = list(itertools.permutations(elements))

# Dizionario per contare le occorrenze di ogni permutazione
counts = Counter()

# Eseguiamo 1.000.000 di permutazioni casuali
num_runs = 1_000_000
for _ in range(num_runs):
    perm = tuple(random.sample(elements, len(elements)))  # Permutazione casuale
    counts[perm] += 1

# Estraiamo le permutazioni e le loro frequenze
x_labels = [''.join(map(str, perm)) for perm in all_permutations]
y_values = [counts[perm] for perm in all_permutations]

#Grafico delle frequenze
plt.figure(figsize=(12, 6))
plt.bar(x_labels, y_values, color='skyblue', edgecolor='black')
plt.xlabel("Permutazioni")
plt.ylabel("Frequenza")
plt.title("Distribuzione delle Permutazioni Casuali")
plt.xticks(rotation=45)
plt.show()
```

```python
# Importa le librerie necessarie
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Dati originali
data = {
    "Dosaggio": ["< 2.00", "2.00 - 3.99", "4.00 - 5.99", "6.00 - 7.99", "8.00 - 9.99",
                 "10.00 - 11.99", "12.00 - 13.99", "14.00 - 15.99", "16.00 - 17.99", "> 18.00"],
    "Negativi": [34, 74, 100, 127, 153, 114, 19, 6, 4, 0],  # Valori osservati negativi
    "Positivi": [2, 6, 10, 12, 23, 55, 75, 41, 30, 15]      # Valori osservati positivi
}

df = pd.DataFrame(data)

# Funzione per determinare la classe predetta in base al cutoff (soglia)
def predici_classe(dosaggio_str, soglia=9.99):
    if dosaggio_str.startswith("<"):
        return "Negativo"
    elif dosaggio_str.startswith(">"):
        return "Positivo"
    else:
        inferiore, superiore = map(float, dosaggio_str.split(" - "))
        return "Negativo" if superiore <= soglia else "Positivo"

# Aggiunge una colonna con la predizione
df["Predetto"] = df["Dosaggio"].apply(predici_classe)

# Calcola TN, FP, FN, TP
TN = df[df["Predetto"] == "Negativo"]["Negativi"].sum()  # Veri Negativi
FP = df[df["Predetto"] == "Negativo"]["Positivi"].sum()  # Falsi Positivi
FN = df[df["Predetto"] == "Positivo"]["Negativi"].sum()  # Falsi Negativi
TP = df[df["Predetto"] == "Positivo"]["Positivi"].sum()  # Veri Positivi

# Matrice di confusione
matrice_confusione = [[TN, FP], [FN, TP]]

print("Matrice di Confusione (Soglia = 9.99):")
print(pd.DataFrame(matrice_confusione,
                   index=["Reale Negativo", "Reale Positivo"],
                   columns=["Predetto Negativo", "Predetto Positivo"]))

# Accuratezza complessiva
ACC = (TP + TN) / (TP + FP + FN + TN)
print("Accuratezza: ", ACC)

# Sensibilità (o Tasso di Veri Positivi)
TPR = TP / (TP + FN)
print("Tasso di Veri Positivi (TPR): ", TPR)

# Specificità (o Tasso di Veri Negativi)
TNR = TN / (TN + FP)
print("Tasso di Veri Negativi (TNR): ", TNR)

# Tasso di Falsi Positivi
FPR = FP / (FP + TN)
print("Tasso di Falsi Positivi (FPR): ", FPR)

# Valore Predittivo Positivo
PPV = TP / (TP + FP)
print("Valore Predittivo Positivo: ", PPV)

# Valore Predittivo Negativo
NPV = TN / (TN + FN)
print("Valore Predittivo Negativo: ", NPV)

# Calcola le somme cumulative
df["Cumulativo_Negativi"] = df["Negativi"].cumsum()
df["Cumulativo_Positivi"] = df["Positivi"].cumsum()

# Calcola FPR e TPR
totale_neg = df["Negativi"].sum()
totale_pos = df["Positivi"].sum()
df["FPR"] = 1 - (df["Cumulativo_Negativi"] / totale_neg)  # 1 - Specificità
df["TPR"] = 1 - (df["Cumulativo_Positivi"] / totale_pos)  # Sensibilità

# Crea una copia numerica per il grafico
fpr_numerico = df["FPR"].values  # Valori numerici FPR
tpr_numerico = df["TPR"].values  # Valori numerici TPR

# Calcolo dell'AUC (area sotto la curva) tramite approssimazione con trapezi
df["AUC"] = (df["FPR"] - df["FPR"].shift(-1)).fillna(0).abs() * df["TPR"]
auc_totale = df["AUC"].sum()

# Calcola la differenza TPR - FPR
df["TPR-FPR"] = df["TPR"] - df["FPR"]

# Arrotonda e formatta i valori per la tabella (virgole invece di punti)
df["FPR"] = df["FPR"].round(3).astype(str).str.replace('.', ',')
df["TPR"] = df["TPR"].round(3).astype(str).str.replace('.', ',')
df["AUC"] = df["AUC"].round(3).astype(str).str.replace('.', ',')
df["TPR-FPR"] = df["TPR-FPR"].round(3).astype(str).str.replace('.', ',')

# Crea la tabella finale
tabella_roc = df[["Dosaggio", "Negativi", "Positivi", "Cumulativo_Negativi", "Cumulativo_Positivi",
                  "FPR", "TPR", "AUC", "TPR-FPR"]]

# Rinomina le colonne per allinearsi all'esempio
tabella_roc.columns = ["Dosaggio", "Osservati Negativi", "Osservati Positivi",
                       "Cumulativi Negativi", "Cumulativi Positivi",
                       "FPR", "TPR", "AUC", "TPR-FPR"]

print(tabella_roc.to_string(index=False))
print()  # Spazio

# Ordina i dati per FPR crescente
df = df.sort_values(by="FPR").reset_index(drop=True)

# Creazione del grafico ROC
plt.figure(figsize=(8, 6))
plt.scatter(fpr_numerico, tpr_numerico, color='green', s=50, label="Punti ROC")
plt.plot(fpr_numerico, tpr_numerico, color='gray', linestyle='--', label="Curva ROC")
plt.plot([0, 1], [0, 1], color='red', linestyle='--', label="Classificatore Casuale")
plt.xlabel('Tasso di Falsi Positivi (FPR)')
plt.ylabel('Tasso di Veri Positivi (TPR)')
plt.title('Curva ROC per il nuovo vaccino nasale')
plt.xlim(0, 1)
plt.ylim(0, 1)
plt.legend()
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.show()
```

```python
#gian
import numpy as np
import pandas as pd
import math

# Configurazione iniziale
np.random.seed(None)
pd.set_option('display.float_format', '{:,.2f} €'.format)

# Dati di input
assets = ['A', 'B', 'C', 'D', 'E']
min_values = np.array([9000, 11000, 5200, 9900, 15000])
max_values = np.array([15000, 12000, 5500, 12300, 17000])
invested_amount = 55000
required_error_perc = 0.05  # 5%

# STEP 1: Calcolo parametri dai valori estremi

print("="*50)
print("STEP 1: CALCOLO PARAMETRI DA VALORI ESTREMI")
print("="*50)
def deviazioneStandard(val_max, val_min):
    values = [val_max, val_min]

    # Calcolo della media
    mean = sum(values) / len(values)

    # Calcolo scarti quadrati
    squared_diffs = [(x - mean) ** 2 for x in values]

    # Deviazione standard popolazione
    std = math.sqrt(sum(squared_diffs) / len(values))
    return std

# Calcoliamo il range totale del portafoglio
total_max = max_values.sum()
total_min = min_values.sum()
media=(total_max+total_min)/2
print()
# (approssimazione per distribuzioni uniformi)
std_dev = deviazioneStandard(total_max, total_min)
absolute_error = media * required_error_perc
required_n = int(np.ceil((3 * std_dev / absolute_error) ** 2))

# Mostriamo i risultati dello Step 1
step1_results = pd.DataFrame({
    'Parametro': ['Valore Massimo Portafoglio',
                 'Valore Minimo Portafoglio',
                 'Deviazione Standard Stimata',
                 'Errore Massimo Richiesto',
                 'Errore Assoluto',
                 'Numero Iterazioni Monte Carlo'],
    'Valore': [f"{total_max:,.2f} €",
              f"{total_min:,.2f} €",
              f"{std_dev:,.2f} €",
              f"{required_error_perc*100}%",
              f"{absolute_error:,.2f} €",
              required_n]
})
print("\nRISULTATI STEP 1:")
print(step1_results.to_string(index=False))

# STEP 2: Generazione valori simulati

print("\n" + "="*50)
print(f"STEP 2: GENERAZIONE DI {required_n} VALORI SIMULATI")
print("="*50)

simulations = []
for i in range(required_n):
    values = np.random.uniform(low=min_values, high=max_values)
    total = np.sum(values)
    simulations.append(values.tolist() + [total])

# Creiamo il DataFrame con tutti i valori
columns = assets + ['Totale Portafoglio']
df = pd.DataFrame(simulations, columns=columns)

# Aggiungiamo righe di valori max e min
df.loc[-2] = max_values.tolist() + [total_max]  # Val max
df.loc[-1] = min_values.tolist() + [total_min]  # Val min
df.index = df.index + 2
df = df.sort_index()

print("\nTABELLA COMPLETA:")
print(df.to_string(col_space=12, justify='center'))

# STEP 3: Analisi di investimento

print("\n" + "="*50)
print("STEP 3: ANALISI DI INVESTIMENTO (55.000 €)")
print("="*50)

portfolio_totals = df.iloc[2:]['Totale Portafoglio']
real_std_dev = portfolio_totals.std()

loss_prob = np.mean(portfolio_totals < invested_amount) * 100
gain_prob_5000 = np.mean(portfolio_totals >= (invested_amount + 5000)) * 100

print("\nRISULTATI FINALI:")
print('Probabilità di perdere soldi: ', f"{loss_prob:.2f}")
print('Probabilità di guadagno ≥ 5000 €: ',f"{gain_prob_5000:.2f}")
```

Abbiamo quindi steso lo stesso codice utilizzando al posto di "KMedoids" l'import "K-Means", questo è compatibile al 100% con Google Colab.

Attenzione: KMeans usa la media del cluster, mentre KMedoids usa un punto reale (medoide), quindi è più robusto agli outlier.

```python
import pandas as pd
import numpy as np
import zipfile
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans #in sostituzione a KMedoids
from sklearn.metrics import silhouette_score
from sklearn.impute import SimpleImputer
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Caricamento del file CSV da ZIP su Github in condivisione
zip_file_path = "bigdata_0/TA_restaurants_curated.csv.zip"
csv_file_name_in_zip = "TA_restaurants_curated.csv"

with zipfile.ZipFile(zip_file_path) as z:
    with z.open(csv_file_name_in_zip) as f:
        df = pd.read_csv(f).head(10000) #imposto un limite di righe da leggere dal file

# Selezione delle colonne numeriche e trasformazione del prezzo
numeric_cols = ['Ranking', 'Rating', 'Number of Reviews']
df['Price Range_Numeric'] = df['Price Range'].map({'$': 1, '$$': 2, '$$$': 3, '$$$$': 4, '$$$$$': 5})
numeric_cols.append('Price Range_Numeric')

# Gestione dei valori mancanti
imputer = SimpleImputer(strategy='mean')
data = imputer.fit_transform(df[numeric_cols])

# Standardizzazione
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)

# Ricerca del k ottimale con silhouette score
silhouette_scores = []
k_range = range(2, 10)

for k in k_range:
    try:
        kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
        labels = kmeans.fit_predict(data_scaled)
        score = silhouette_score(data_scaled, labels)
        silhouette_scores.append(score)
        print(f"k={k} -> silhouette score: {score:.4f}")
    except Exception as e:
        print(f"Errore con k={k}: {e}")
        silhouette_scores.append(-1)

# Selezione del k migliore
optimal_k = k_range[np.argmax(silhouette_scores)]
print(f"\n✅ Numero ottimale di cluster: {optimal_k}")

# Clustering finale con KMeans al posto di KMedoids
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
df['Cluster'] = kmeans.fit_predict(data_scaled)

# Visualizzazione con PCA
pca = PCA(n_components=2)
components = pca.fit_transform(data_scaled)
df['PC1'], df['PC2'] = components[:, 0], components[:, 1]

plt.figure(figsize=(10, 6))
scatter = plt.scatter(df['PC1'], df['PC2'], c=df['Cluster'], cmap='tab10', s=50, alpha=0.7)
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("Cluster dei ristoranti (K-Means + PCA)")
plt.colorbar(scatter, label="Cluster")
plt.grid(True)
plt.show()

# Statistiche per cluster
print("\nStatistiche per cluster:")
print(df.groupby('Cluster')[numeric_cols].mean())
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.naive_bayes import CategoricalNB
from sklearn.pipeline import make_pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

# Dati slide 9
data = pd.DataFrame({
    'keyword': ['offerta', 'offerta', 'offerta', 'offerta', 'money', 'money', 'money', 'viagra', 'viagra', 'viagra'],
    'lunghezza': ['lunga', 'lunga', 'corta', 'media', 'media', 'lunga', 'corta', 'lunga', 'lunga', 'media'],
    'intestazione': ['si', 'no', 'no', 'si', 'no', 'si', 'no', 'si', 'no', 'no'],
    'valida': ['si', 'si', 'no', 'si', 'no', 'no', 'no', 'no', 'si', 'no']
})

# Mappiamo la colonna target: si = 1 (valida), no = 0 (spam)
data['valida'] = data['valida'].map({'si': 1, 'no': 0})

# ----------------------------
# Calcolo manuale P(x|c)
# ----------------------------
def calcola_probabilita_condizionata(colonna):
    valori = data[colonna].unique()
    classi = [1, 0]  # 1 = valida, 0 = spam
    probabilita = {}

    for val in valori:
        probabilita[val] = {}
        for c in classi:
            numeratore = len(data[(data[colonna] == val) & (data['valida'] == c)])
            denominatore = len(data[data['valida'] == c])
            probabilita[val]['valida' if c == 1 else 'spam'] = numeratore / denominatore if denominatore > 0 else 0

    return pd.DataFrame(probabilita).T

# Calcoliamo le tabelle per ciascuna feature
p_keyword = calcola_probabilita_condizionata('keyword')
p_lunghezza = calcola_probabilita_condizionata('lunghezza')
p_intestazione = calcola_probabilita_condizionata('intestazione')

# Visualizzazione
print("📧 P(keyword | classe):\n", p_keyword, "\n")
print("📏 P(lunghezza | classe):\n", p_lunghezza, "\n")
print("🧾 P(intestazione | classe):\n", p_intestazione, "\n")

# Probabilità a priori
prior_valida = len(data[data['valida'] == 1]) / len(data)
prior_spam = 1 - prior_valida
print(f"🔢 P(email valida) = {prior_valida:.2f}")
print(f"🔢 P(email spam) = {prior_spam:.2f}")
```
